Start to integrate three dataframe into one.

```python
import pandas as pd
from urllib.request import urlopen
import numpy as np
```

Get the three datasets from github:

```python
DF=pd.read_pickle(urlopen("https://github.com/tmq2uw/McQueenyRepo/raw/master/pftest.pkl"),compression=None)
DF2=pd.read_pickle(urlopen("https://github.com/mav95-evans/Data/raw/master/giiIndex2018.pkl"),compression=None)
DF3=pd.read_pickle(urlopen("https://github.com/elsie216/542_MyData/raw/master/GDPpercap1.pkl"),compression=None)
```

Unify the column names:

```python
DF2.columns=["Country", "Giilndex2018"]
```

Start by merging DF and DF2, and rename the merged new dataframe as "Newtable":

```python
DF.merge(DF2,how='outer',indicator=True)
```

```python
Newtable=DF.merge(DF2,how='outer',indicator=True)
```

```python
Newtable
```

Then integrate DF3 into the previous new dataframe "Newtable":

```python
Newtable.merge(DF3, how='outer',indicator='exists')
```

Then we start to clean the integrated dataframe:

```python
new_df=Newtable.merge(DF3, how='outer',indicator='exists')
```

```python
# Dropping all rows that do not have _merg==Both
indexNames = new_df[new_df['exists']!='both'].index
 
# Delete these row indexes from dataFrame
new_df.drop(indexNames , inplace=True)
new_df.reset_index(drop=True,inplace=True)
```

```python
new_df
```

```python
type(new_df)
```

Drop the columns we don't need:

```python
groupdata1=new_df.drop(['_merge'], axis=1)
```

```python
groupdata1
```

```python
groupdata=groupdata1.drop(['exists'], axis=1)
```

```python
groupdata
```

Check the types of each variable(column):

```python
groupdata.dtypes
```

Convert the variable "GiiIndex2018" from object to numeric

```python
groupdata['Giilndex2018']=pd.to_numeric(groupdata['Giilndex2018'])
```

Save the dataset as RDS to be used in R:

```python
from rpy2.robjects import pandas2ri
pandas2ri.activate()

from rpy2.robjects.packages import importr

base = importr('base')
base.saveRDS(groupdata,file="TEAMdata.RDS")
```
